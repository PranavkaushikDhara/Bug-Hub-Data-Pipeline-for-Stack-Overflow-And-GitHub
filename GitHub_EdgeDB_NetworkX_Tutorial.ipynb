```python
# Common API headers
headers = {
    "Accept": "application/vnd.github+json",
    "access_token": "ADD-YOUR-ACCESS-TOKEN",
    "Git_Username":"ADD YOUR-GITHUB USERNAME"
}
```

```python
# Define a dictionary of owners and their repositories
owner_repos = {
    'angular': ['angular-cli']
}
```

```python
print(headers["access_token"]);
print(headers["Git_Username"])
def fetch_url(owner, repo):
    return f"https://"+headers["Git_Username"]+":"+headers["access_token"]+f"@api.github.com/repos/{owner}/{repo}"
```

```python
def fetch_stars(owner, repo):
    url = fetch_url(owner,repo)
    response = requests.get(url, headers=headers)
    data = response.json()
    return data.get('stargazers_count', 0)
```

```python
def update_repository_stars(repo_name, stars):
    client.query("""
        UPDATE Repository
        FILTER .name = <str>$name
        SET {stars := <int32>$stars}
    """, name=repo_name, stars=stars)
    
def insert_repository_and_stars(repo_name, stars):
    client.query("""
        INSERT Repository{
        name := <str>$name,
        stars := <int32>$stars}
    """, name=repo_name, stars=stars)
```

```python
#defining per page value to decide the amount of data to fetch
page = 1
per_page = 2

#defining since date

since_date = (datetime.date.today() - datetime.timedelta(days=730)).isoformat()
```

```python
def fetch_and_insert_forks(owner, repo, page, per_page, since_date):        
    repo_forks_ids=[]
    repo_forks_names=[]
    repo_forks_owners=[]
    url = fetch_url(owner, repo)
    while True:
        response = requests.get(f"{url}/forks", headers=headers,params={"since": since_date, "page": page})
        for obj in response.json():
            repo_forks_ids.append(obj['id'])
            repo_forks_names.append(obj['full_name'])
            repo_forks_owners.append(obj['owner']['login'])

        if not response.ok or len(response.json()) == 0 :
            break

        page+=1
    
    page = 1

    for i in range(len(repo_forks_ids)):
        client.query("""
            INSERT Forks {
                ID := <int64>$ID,
                name :=<str>$name,
                in_repo := (
                select Repository
                filter
                    .name = <str>$repo
                limit 1
            ),
            has_owner := (
            select Contributors filter.name =<str>$owner
            )
            } unless conflict on .name
        """, ID=repo_forks_ids[i],name=repo_forks_names[i],repo=repo,owner=repo_forks_owners[i]);

```

```python
def fetch_and_insert_contributors(owner, repo, page, per_page, since_date):
    repo_contributors=[]
    repo_contributions=[]
    url = fetch_url(owner, repo)
    while True:
        response = requests.get(f"{url}/contributors", headers=headers,params={"since": since_date, "page": page})
        for obj in response.json():
            repo_contributors.append(obj['login'])
            repo_contributions.append(obj['contributions'])

        if not response.ok or len(response.json()) == 0:
            break

        page+=1
    
    page = 1

    for i in range(len(repo_contributors)):
        client.query("""
            INSERT Contributors {
                name := <str>$name,
                contributions :=<int32>$contributions,
                in_repo := (
                select Repository
                filter
                    .name = <str>$repo
                limit 1
            )
            } unless conflict on .name
        """, name=repo_contributors[i],contributions=repo_contributions[i],repo=repo);
```

```python
def fetch_and_insert_commits(owner, repo, page, per_page, since_date):
    repo_commits = []
    repo_authors= []
    repo_committers= []
    
    url = fetch_url(owner, repo)

    # loop until all pages have been retrieved
    while True:
        # make the API request
        response = requests.get(f"{url}/commits", params={"since": since_date, "page": page}, headers=headers)
        if not response.ok or len(response.json()) == 0:
            break

        # append the current page's results to the list of commits
        for obj in response.json():
            repo_commits.append(obj['sha'])
            repo_authors.append(obj['commit']['author']['name'])
            repo_committers.append(obj['commit']['committer']['name'])

        #increment the page number
        page += 1
    
    page=1

    for i in range(len(repo_commits)):
        client.query("""
            INSERT Commits {
                sha := <str>$sha,
                author :=<str>$author,
                committer :=<str>$committer,
                has_author :=(
                select Contributors
                filter
                    .name = <str>$author
                limit 1
            ),
            has_committer :=(
                select Contributors
                filter
                    .name = <str>$committer
                limit 1
            ),
                in_repo := (
                select Repository
                filter
                    .name = <str>$repo
                limit 1
            )
            } unless conflict on .sha
        """, sha=repo_commits[i],author=repo_authors[i],committer=repo_committers[i],repo=repo);

```

```python
def fetch_and_insert_labels(owner, repo, page, per_page, since_date):
    repo_labels_ids = []
    repo_labels = []

    url = fetch_url(owner, repo)
    # loop until all pages have been retrieved
    while True:
        # make the API request
        response = requests.get(f"{url}/labels", params={"since": since_date, "page": page}, headers=headers)
        if not response.ok or len(response.json()) == 0:
            break

        # append the current page's results to the list of commits
        for obj in response.json():
            repo_labels.append(obj['name']);
            repo_labels_ids.append(obj['id']);


        #increment the page number
        page += 1

    page = 1
    for i in range(len(repo_labels)):
        client.query("""
            INSERT Labels {
                ID := <int64>$ID,
                name :=<str>$label,
                in_repo := (
                select Repository
                filter
                    .name = <str>$repo
                limit 1
            )
            } unless conflict on .ID
        """, ID=repo_labels_ids[i],label=repo_labels[i],repo=repo);
```

```python
def fetch_and_insert_branches(owner, repo, page, per_page, since_date):
    repo_branches=[]
    url = fetch_url(owner, repo)
    print(url)
    while True:
        response = requests.get(f"{url}/branches", headers=headers,params={"since": since_date, "page": page})
        for obj in response.json():
            repo_branches.append(obj['name'])


        if not response.ok or len(response.json()) == 0:
            break

        page+=1
        
    page = 1

    for i in range(len(repo_branches)):
        client.query("""
    INSERT Branches {
        name :=<str>$name,
        in_repo := (
        select Repository
        filter
            .name = <str>$repo
        limit 1
    )
    } unless conflict on .name
""",name=repo_branches[i],repo=repo);
```

```python
def fetch_and_insert_releases(owner, repo, page, per_page, since_date):
    repo_releases=[]
    url = fetch_url(owner, repo)
    while True:
        response = requests.get(f"{url}/releases", headers=headers,params={"since": since_date, "page": page})
        for obj in response.json():
            repo_releases.append(obj['tag_name'])

        if not response.ok or len(response.json()) == 0:
            break

        page+=1
    
    page = 1

    for i in range(len(repo_releases)):
        client.query("""
    INSERT Releases {
        name :=<str>$name,
        in_repo := (
        select Repository
        filter
            .name = <str>$repo
        limit 1
    )
    } unless conflict on .name
""",name=repo_releases[i],repo=repo);
```

```python
def fetch_and_insert_issues(owner, repo, page, per_page, since_date):
    repo_issues_numbers=[]
    repo_issues_created=[]
    repo_issues_closed=[]
    
    url = fetch_url(owner, repo)
    while True:
        response = requests.get(f"{url}/issues", headers=headers,params={"since": since_date, "page": page,"state":"all"})
        for obj in response.json():
            print(response.json())
            repo_issues_numbers.append(obj["number"])
            repo_issues_created.append(obj["created_at"])
            repo_issues_closed.append(obj['closed_at'])

        if not response.ok or len(response.json()) == 0:
            break

        page+=1

    for i in range(len(repo_issues_numbers)):
        cdate=repo_issues_created[i];
        cldate=repo_issues_closed[i];
        created=datetime.datetime.fromisoformat(cdate)
        client.query("""
    INSERT Issues {
        number :=<int32>$number,
        created_at:=<datetime>$created,
        closed_at:=<str>$closed,
        week_of_year:= (
        datetime_get(<datetime>$created,"week")
        ),
        in_repo := (
        select Repository
        filter
            .name = <str>$repo
        limit 1
    )
    } unless conflict on .number
""",number=repo_issues_numbers[i],repo=repo,created=datetime.datetime.fromisoformat(str(repo_issues_created[i])),
                     closed=(str(repo_issues_closed[i])));
```

```python
def fetch_and_insert_pull_requests(owner, repo, page, per_page, since_date):
    repo_pulls=[]
    repo_pulls_users=[]
    repo_pulls_states=[]
    url = fetch_url(owner, repo)
    
    while True:
        response = requests.get(f"{url}/pulls", headers=headers,params={"since": since_date, "page": page,"state":"all"})
        print(response.json())
        for obj in response.json():
            repo_pulls.append(obj["number"])
            repo_pulls_users.append(obj['user']['login'])
            repo_pulls_states.append(obj["state"])


        if not response.ok or len(response.json()) == 0:
            break

        page+=1

    page =1
    
    for i in range(len(repo_pulls)):
        client.query("""
    INSERT Pulls {
        number :=<int32>$number,
        state:=<str>$state,
        has_owner:=(
        select Contributors filter.name =<str>$owner
        ),
        in_repo := (
        select Repository
        filter
            .name = <str>$repo
        limit 1
    )
    } unless conflict on .number
""",number=repo_pulls[i],repo=repo,state=repo_pulls_states[i],owner=repo_pulls_users[i]);

```

```python
#NetworkX Github

import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()

# add nodes
G.add_node('Git')
G.add_node('GitHub')
G.add_node('Project')
G.add_node('Repository')
G.add_node('Fork')
G.add_node('Contributor')
G.add_node('Branch')
G.add_node('Issue')
G.add_node('Pull')
G.add_node('Label')
G.add_node('Commit')
G.add_node("Release")
G.add_node("Tag")
G.add_node("Topic")

G.add_edges_from([('Git', 'GitHub',{'label':'used_by'}), ('GitHub', 'Repository',{'label':'has_repository'}),('GitHub', 'Project',{'label':'has_project'}),
                  ('GitHub', 'Contributor',{'label':'has_User'})
                 ,('Repository', 'Release',{'label':'has_release'}),
                  ('Repository', 'Fork',{'label':'has_fork'}),
                  ('Repository', 'Tag',{"label":"has_tag"}),('Repository', 'Branch',{"label":"has_branch"}), ('Repository', 'Issue',{"label":"has_issue"}), ('Repository', 'Pull',{"label":"has_pull"}), ('Issue', 'Label',{"label":"has_label"}),
                 ("Topic","Repository",{"label":"in_repo"}), ("Repository","Topic",{"label":"has_topic"}),("Release","Tag",{"label":"has_tag"}),("Project","Repository",{"label":"project_has_repo"}),('Repository',"Contributor",{"label":"has_contributor"}), ('Branch', 'Commit',{"label":"branch_has_commit"}), ('Fork', 'Repository',{"label":"is_child_to"}),('Pull',"Commit",{"label":"has_commit"})])

pos = nx.spring_layout(G, seed=120,k=2)

nx.draw(G, pos, with_labels=True, font_size=10, node_size=1000, node_color='lightblue', edge_color='gray', font_weight='bold')

# Add edge labels
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8, font_weight='bold')

plt.title("Graph data model for GitHub")
plt.show()
```
